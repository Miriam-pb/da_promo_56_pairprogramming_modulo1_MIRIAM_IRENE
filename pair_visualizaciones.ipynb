¿Cómo se distribuye la tasa de natalidad en el conjunto de datos? ¿Hay algún patrón evidente?

```python
sns.histplot(x='birth_rate',
             data= df,
             color= 'violet',
             kde=True
             );
plt.title('tasa natalidad')
plt.xlabel('nacimientos cada 1000 habitantes')
plt.ylabel('continente')
```

```python
sns.scatterplot(x = "birth_rate", 
                y = "continent", 
                data = df, 
                size = 5, 
                marker= "o", 
                color = "violet")
plt.title('tasa natalidad')
plt.xlabel('nacimientos cada 1000 habitantes')
plt.ylabel('continente')
```

conclusión: la mayoría de paises tienen una tasa de natalidad baja, la tasa de natalidad no esta relacionada con la poblacion, esta relacionada con el desarrollo del pais. Los paises mas desarrollados tienen una tasa más baja

¿Existe una relación aparente entre la tasa de mortalidad infantil y la esperanza de vida? Proporciona una breve descripción de los patrones que observes.

```python
sns.scatterplot(x = "infant_mortality", 
                y = "life_expectancy", 
                data = df, 
                size = 5, 
                marker= "o", 
                color = "red")
plt.title('relación esperanza de vida con mortalidad infantil')
plt.xlabel('mortalidad infantil')
plt.ylabel('esperanza de vida')
```

conclusión: a mayor esperanza de vida menor tasa de mortalidad infantil

¿Existen relaciones visibles entre la tasa de natalidad y las siguientes variables: tasa de mortalidad infantil, esperanza de vida, densidad de población y población urbana? Usa un subplot. Describe cualquier patrón que encuentres.

```python
df.columns
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 4, figsize = (20, 5), sharex= True)

sns.scatterplot(x = "birth_rate", 
                y = "infant_mortality", 
                data = df, 
                size = 5, 
                marker= "o", 
                color = "red",
                ax = axes[0])
axes[0].set_title("relacion tasa natalidad y mortalidad infantil")
axes[0].set_xlabel('tasa natalidad')
axes[0].set_ylabel('mortalidad infantil')

sns.scatterplot(x = "birth_rate",
                y = "life_expectancy", 
                data = df, 
                size = 5, 
                marker= "o", 
                color = "orange",
                ax = axes[1])

axes[1].set_title("relacion tasa natalidad y esperanza de vida")
axes[1].set_xlabel('tasa natalidad')
axes[1].set_ylabel('esperanza de vida')

sns.scatterplot(x = "birth_rate",
                y = "density", 
                data = df, 
                size = 5, 
                marker= "o", 
                color = "green",
                ax = axes[2])

axes[2].set_title("relacion tasa natalidad y densidad")
axes[2].set_xlabel('tasa natalidad')
axes[2].set_ylabel('densidad')

sns.scatterplot(x = "birth_rate",
                y = "population", 
                data = df, 
                size = 5, 
                marker= "o", 
                color = "pink",
                ax = axes[3])

axes[3].set_title("relacion tasa natalidad y poblacion")
axes[3].set_xlabel('tasa natalidad')
axes[3].set_ylabel('poblacion');
```

**conclusiones:**
- a mayor tasa de natalidad mayor mortalidad infantil, los paises mas desarrollados son los que menor tasa de natalidad y menor mortalidad infantil tienen.
- a mayor esperanza de vida menor tasa de natalidad, los paises mas desarrollados son los que tienen mas esperanza de vida y menos natalidad.
- la densidad de poblacion es lineal por lo que no depende de la tasa de natalidad.
- la poblacion y la tasa de natalidad no tiene relacion.
